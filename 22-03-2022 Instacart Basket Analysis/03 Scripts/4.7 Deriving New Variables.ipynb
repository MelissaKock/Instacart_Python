# If-Statements - User defined functions

```python
# Creating data subset with 1mil rows

df = ords_prods_merged[:1000000]
```

```python
df.shape
```

```python
# Define price lable row

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
    
```

```python
# Apply price lable function

df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
# Max value of a product

df['prices'].max()
```

# If-Statements  - For-loops

```python
# Orders per day of the week

ords_prods_merged['order_dow'].value_counts(dropna = False)
```

```python
ords_prods_merged.head()
```

```python
# Loop Function

result = []

for value in ords_prods_merged["order_dow"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
# Create new column ('Result')

ords_prods_merged['busiest_day'] = result
```

```python
ords_prods_merged['busiest_day'].value_counts(dropna = False)
```

```python
# 10am has highest orders/hours to 3am with least orders
```

```python
# Loop

result_hod = []

for value in ords_prods_merged["order_hour_of_day"]:
  if value > 8 & value <= 18:
    result_hod.append("Most_orders")
  elif value <= 5:
    result_hod.append("Fewest_orders")
  else:
    result_hod.append("Average_orders")
```

```python
# New column for 'result_hod'

ords_prods_merged['busiest_period_of_day'] = result_hod
```
